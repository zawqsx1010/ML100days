#### 截取部分資料

```python
app_train1.iloc[4] # 擷取ID第四列之資料
```

```python
app_train1.iloc[4,5] # 擷取ID第四列第6行資料
```

```python
app_train1.iloc[[4,5,6]] # 擷取ID第四到六列資料(多行輸出要加上[])
```

```python
app_train1.iloc[[4,5,6],[0,3]] # 擷取ID第四到六列資料的第1與第4變數
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
app_train1.columns # 資料欄位名稱,也可利用column name擷取資料
```

```python
app_train1.loc[app_train1["NAME_CONTRACT_STATUS"]=="Active"] # 利用擷取擁有特定特徵文字資料
```

```python
app_train1.nlargest(3,"MONTHS_BALANCE")    #查詢MONTHS_BALANCE欄位中數值前3大的
app_train1.nsmallest(3,"MONTHS_BALANCE")   #查詢MONTHS_BALANCE欄位中數值前3小的
```
