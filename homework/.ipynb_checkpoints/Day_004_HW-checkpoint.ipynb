## 題目:請通過 HomeCredit_columns_description.csv，了解各個欄位的意義

## Ans:

由以下資料的顯示，大致上可知道"HomeCredit_columns_description.csv"資料中儲存著多個資料檔的說明，總共資料包含219列以及5行，列代表著每一個不同的資料檔，其中每列都包含了5個變數，其中有一空白欄位由"Unnamed: 0"呈現，猜測為資料的編號；其餘變數Table可能為資料檔名稱、變數Row代表此資料檔各列代表的意涵、變數Description則為對資料檔的解釋、變數Special則像是提醒資料檔特別之處。

```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = r'D:\GitHub\ML100Days\homework\data'
```

```python
f_app = os.path.join(dir_data, 'HomeCredit_columns_description.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app,sep=',',engine='python')   #資料由","分割
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

須知資料的行列數可利用物件導向中的shape幫我們得知，"HomeCredit_columns_description.csv"資料當中擁有219列，且每列包含5個的變數。


```python
app_train.shape # 顯示行列數
```

#### 列出所有欄位

```python
print(app_train) # 列出所有欄位
```

```python
app_train.head() # 列出首5列
```

```python
app_train.tail(n=3) # 列出末三列
```

#### 截取部分資料

```python
app_train.loc[app_train['Special'].notna()] # 選取'Special'中非遺失值的資料
```

```python
app_train.iloc[4] # 擷取ID第四列之資料
```

```python
app_train.iloc[4,3] # 擷取ID第四列第4行資料
```

```python
app_train.iloc[[4,5,6]] # 擷取ID第四到六列資料(多行輸出要加上[])
```

```python
app_train.iloc[[4,5,6],[0,3]] # 擷取ID第四到六列資料的第1與第4變數
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
app_train.columns # 資料欄位名稱,也可利用column name擷取資料
```

```python
app_train.loc[app_train["Table"]=="installments_payments.csv"] # 利用擷取擁有特定特徵文字資料
```

```python
app_train.nlargest(3,"Unnamed: 0")    #查詢MONTHS_BALANCE欄位中數值前3大的
app_train.nsmallest(3,"Unnamed: 0")   #查詢MONTHS_BALANCE欄位中數值前3小的
```
